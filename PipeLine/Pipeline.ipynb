```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/train (1).csv")
df
```

```python
df = df.drop(['PassengerId','Name',"Ticket","Cabin"],1)
```

```python
x = df.drop(['Survived'],1)
y = df.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
t1 = ColumnTransformer(transformers=[
    ("Age_mean",SimpleImputer(),[2]),
    ("Embarked_mostfrq",SimpleImputer(strategy="most_frequent"),[6])
],remainder='passthrough')
```

```python
t2 = ColumnTransformer(transformers=[
    ('Ohe_Embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
t3 = ColumnTransformer(transformers = [
    ('Stdscalser',StandardScaler(),slice(0,8))
])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
t4 = DecisionTreeClassifier()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('t1',t1),
    ('t2',t2),
    ('t3',t3),
    ('t4',t4)
])
```

```python
pipe.named_steps['t2'].transformers
```

```python
pipe.fit(x_train,y_train)    #if we not training model then it will be fit_transform that means for preprocessing.
```

```python
yprd = pipe.predict(x_test)
```

```python
yprd
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yprd)
```

Cross Validation Using Pipeline

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,x_train,y_train,cv = 5,scoring='accuracy').mean()
```

GridSearch Using Pipeline

```python
params = {
    't4__max_depth':[1,2,3,4,5,6,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe,params,cv = 5, scoring = 'accuracy')
grid.fit(x_train,y_train)
```

```python
grid.best_score_
```
