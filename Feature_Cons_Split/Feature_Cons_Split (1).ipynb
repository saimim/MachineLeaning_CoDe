```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/content/train (3).csv',usecols=('Age', 'Parch','SibSp','Survived','Pclass'))
```

```python
df
```

```python
x = df.drop(['Survived'],axis = 1)
y = df.Survived
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming x is your feature matrix and y is your target variable

# Define the imputer strategy, e.g., mean imputation
imputer = SimpleImputer(strategy='mean')

# Create a pipeline with imputation and Logistic Regression
pipeline = Pipeline(steps=[('imputer', imputer),
                           ('classifier', LogisticRegression())])

# Calculate mean cross-validated accuracy after imputation
mean_accuracy = np.mean(cross_val_score(pipeline, x, y, scoring='accuracy', cv=20))

print(f"Mean cross-validated accuracy after imputation: {mean_accuracy}")
```

Applying Feature Construction

```python
x['Family_size'] = x['Parch']+x['SibSp']+1
```

```python
x
```

```python
def ffun(num):
  if num == 1:
    return 0
  elif num>1 and num<=4:
    return 1
  else:
    return 2
```

```python
ffun(4)
```

```python
x['Family_type'] = x['Family_size'].apply(ffun)
```

```python
x.head()
```

```python
x = x.drop(['SibSp','Parch','Family_size'],axis = 1)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming x is your feature matrix and y is your target variable

# Define the imputer strategy, e.g., mean imputation
imputer = SimpleImputer(strategy='mean')

# Create a pipeline with imputation and Logistic Regression
pipeline = Pipeline(steps=[('imputer', imputer),
                           ('classifier', LogisticRegression())])

# Calculate mean cross-validated accuracy after imputation
mean_accuracy = np.mean(cross_val_score(pipeline, x, y, scoring='accuracy', cv=20))

print(f"Mean cross-validated accuracy after imputation: {mean_accuracy}")
```
