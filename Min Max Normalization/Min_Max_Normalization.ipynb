```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/wine_data.csv",header = None,usecols = [0,1,2])
df.columns=['class label','Alcohol','Malic acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
sns.scatterplot(x = df['Alcohol'],y = df['Malic acid'],hue=df['class label'])
```

```python
 x = df.drop(['class label'],1)
 y = df['class label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .2,random_state = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
c = ['Alcohol','Malic acid']
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled,columns = c)
```

```python
x_train_scaled
```

```python
x_train_scaled.describe()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize=(12,5))
ax1.scatter(x = x_train['Alcohol'],y = x_train['Malic acid'], c = y_train)
ax1.set_title("Before Scaling")
ax2.scatter(x = x_train_scaled['Alcohol'], y = x_train_scaled['Malic acid'], c = y_train)
ax2.set_title('After Scaling')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)
sns.kdeplot(x_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'], ax=ax2)
plt.show()
```
