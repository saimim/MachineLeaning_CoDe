```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/CSE322(ML & DM L)/Churn_Modelling-m.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.isnull().mean()*100
```

```python
df['Gender'] = le.fit_transform(df.Gender)
```

```python
df['Gender'].value_counts().sort_values(ascending = False)
```

```python
df['Gender'] = [np.nan if x == 2 else x for x in df['Gender']]
```

```python
df
```

```python
df['Gender'].value_counts()
```

```python
imputer = IterativeImputer(random_state = 100)
```

```python
df_train = pd.DataFrame(df[['Exited', 'Gender', 'Balance']])
```

```python
df_train
```

```python
df_train = imputer.fit_transform(df_train)
```

```python
df_train
```

```python
df1 = pd.DataFrame(df_train[:,1].round(), columns=['Gender_en'])
```

```python
df1
```

```python
df = pd.concat([df,df1],axis = 1)
```

```python
df
```

```python
df['Gender_en'] = ['Female' if x == 0.0 else 'Male' for x in df['Gender_en']]
```

```python
df.columns
```

```python
df = df[['RowNumber', 'CustomerId', 'Surname', 'CreditScore', 'Geography','Gender_en', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary', 'Exited']]
```
