```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CSE322(ML & DM L)/Titanic-Toy.csv')
df
```

```python
df.isnull().mean()
```

```python
x = df.drop(['Survived'], axis = 1)
y = df.Survived
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2)
```

```python
x_train['Age_imputed'] = x_train['Age']
x_test['Age_imputed'] = x_test['Age']

x_train['Fare_imputed'] = x_train['Fare']
x_test['Fare_imputed'] = x_test['Fare']
```

```python
x_train['Age_imputed'][x_train['Age_imputed'].isnull()] = x_train['Age'].dropna().sample(x_train['Age'].isnull().sum(),random_state = 1).values
x_test['Age_imputed'][x_test['Age_imputed'].isnull()] = x_test['Age'].dropna().sample(x_test['Age'].isnull().sum(),random_state = 1).values
```

```python
x_train['Age'].dropna().sample(x_train['Age'].isnull().sum()).values
```

```python
x_train['Age'].isnull().sum()
```

```python
x_train
```

```python
x_train['Fare_imputed'][x_train['Fare_imputed'].isnull()] = x_train['Fare'].dropna().sample(x_train['Fare'].isnull().sum(),random_state = 1).values
```

```python
x_test['Fare_imputed'][x_test['Fare_imputed'].isnull()] = x_test['Fare'].dropna().sample(x_test['Fare'].isnull().sum(),random_state = 1).values
```

```python
sns.distplot(x_train['Age'],label = 'Orginal',hist = False)
sns.distplot(x_train['Age_imputed'],label = 'Imputed',hist = False)

plt.legend()
plt.show()
```

```python
sns.distplot(x_train['Fare'],label = 'Orginal',hist = False)
sns.distplot(x_train['Fare_imputed'],label = 'Imputed',hist = False)

plt.legend()
plt.show()
```

```python
print('Orginal variable variance: ', x_train['Age'].var())
print('Variance after random imputation: ', x_train['Age_imputed'].var())

print('Orginal variable variance: ', x_train['Fare'].var())
print('Variance after random imputation: ', x_train['Fare_imputed'].var())
```

```python
x = data
y = data.SalePrice
```

```python
x_train['GarageQual_imputed'] = x_train['GarageQual']
x_test['GarageQual_imputed'] = x_test['GarageQual']

x_train['FireplaceQu_imputed'] = x_train['FireplaceQu']
x_test['FireplaceQu_imputed'] = x_test['FireplaceQu']
```

```python
x_train.sample(5)
```

```python
x_train['GarageQual_imputed'][x_train['GarageQual_imputed'].isnull()] = x_train['GarageQual'].dropna().sample(x_train['GarageQual'].isnull().sum(),random_state = 1).values
x_test['GarageQual_imputed'][x_test['GarageQual_imputed'].isnull()] = x_test['GarageQual'].dropna().sample(x_test['GarageQual'].isnull().sum(),random_state = 1).values


x_train['FireplaceQu_imputed'][x_train['FireplaceQu_imputed'].isnull()] = x_train['FireplaceQu'].dropna().sample(x_train['FireplaceQu'].isnull().sum(),random_state = 1).values
x_test['FireplaceQu_imputed'][x_test['FireplaceQu_imputed'].isnull()] = x_test['FireplaceQu'].dropna().sample(x_test['FireplaceQu'].isnull().sum(),random_state = 1).values
```
