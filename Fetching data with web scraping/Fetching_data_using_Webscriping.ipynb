```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
Salaries = []
for i in company:
  name.append(i.find('h2').text.strip())   #strip user for avoiding tab
  rating.append(i.find_all('span', class_='companyCardWrapper__companyRatingValue')[0].text.strip())
  Salaries.append(i.find_all('span', class_='companyCardWrapper__ActionCount')[0].text.strip())

d = {'name' : name,
     'rating': rating,
     'Salaries': Salaries}
df = pd.DataFrame(d)
df
```
