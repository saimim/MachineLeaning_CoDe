```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv("/content/orders.csv")
time = pd.read_csv("/content/messages.csv")
```

```python
date
```

```python
date.info()
```

Change Date Time Type

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

Extract Year

```python
date['date_year'] = date['date'].dt.year
date
```

Extract Month

```python
date['date_month'] = date['date'].dt.month
date
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date
```

Extract Day

```python
date['day'] =date['date'].dt.day
```

```python
date
```

Day which week


```python
date['day_of_week'] = date['date'].dt.dayofweek
```

```python
date
```

```python
date['day_name'] = date['date'].dt.day_name()
date
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['day_name'].isin(['Friday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Week of the year

```python
date['date_week'] = date['date'].dt.week
date
```

Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
